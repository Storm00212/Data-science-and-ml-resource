```python
# !pip install llama-cpp-python

from llama_cpp import Llama

llm = Llama.from_pretrained(
	repo_id="joshnader/Meta-Llama-3.1-8B-Instruct-Q4_K_M-GGUF",
	filename="meta-llama-3.1-8b-instruct-q4_k_m.gguf",
)
```
